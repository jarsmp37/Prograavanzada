```python
df=df.rename(columns={"age":"Edad","gender":"Genero","smoking_status":"Fumador","bmi":"Indice_masa_corp",
                      'blood_pressure':"Presion_Sanguinea", 'glucose_levels':"Nivel_glucosa", 'condition':"Condicion"})
df
```

```python
#borrar columna
df=df.drop(columns=["id"])
df
```

```python
for columna in df.columns:
    print(f"La columna {columna} tiene de datos")
    print(df[columna].unique())
    print("")
```

```python
dic1={'male':"Hombre", 'female':"Mujer",'Non-Smoker':"No_fumador", 'Smoker':"Fumador",'Pneumonia':"Neumonia",
        'Diabetic':"Diabetico"}
```

```python
df["Genero"]=df["Genero"].replace(dic1)
df
```

```python
for columna in df.columns:
    df[columna]=df[columna].replace(dic1)
```

```python
import plotly.express as px

conteo_condicion = df['Condicion'].value_counts().reset_index()
conteo_condicion.columns= ["Enfermedad", "Count"]

fig = px.bar(conteo_condicion, x="Enfermedad", y="Count", title="Enferemedades en pacientes", color="Enfermedad")

fig.show()
```

```python
df.columns
```

```python
import plotly.express as px

conteo_condicion_genero=df.groupby(["Condicion", "Genero"]).size().reset_index(name="Count")
conteo_condicion_genero.columns=["Enfermedad","Genero","Count"]
fig=px.bar(conteo_condicion_genero,x="Enfermedad",y="Count",color="Enfermedad",facet_col="Genero",title="Enfermedades por GÃ©nero")

fig.show()
```

```python
import plotly.express as px

conteo_condicion_genero=df.groupby(["Condicion", "Fumador"]).size().reset_index(name="Count")
conteo_condicion_genero.columns=["Enfermedad","Fumador","Count"]
fig=px.bar(conteo_condicion_genero,x="Enfermedad",y="Count",color="Enfermedad",facet_col="Fumador",title="Enfermedades por GÃ©nero")

fig.show()
```
